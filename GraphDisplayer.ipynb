```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
```

```python
# adding nodes

G.add_node(1) # add a node named 1
G.add_nodes_from([2, 3]) # add nodes named 2 and 3
G.add_nodes_from([
    (4, {"color": "red"}), # add nodes of form (node, node_attribute_dict)
    (5, {"color": "green"})
])
```

```python
# adding edges

G.add_edge(1, 2) # add edges one at a time
e = (2, 3)
G.add_edge(*e) # unpacking edge tuple
G.add_edges_from([(1, 2), (1, 3)]) # add a list of edges
```

```python
# clear graph and add new nodes and edges

G.clear() # remove all edges and nodes
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
# print edges and nodes

print(G.number_of_nodes())
print(G.number_of_edges())
print(list(G.nodes))
print(list(G.edges))
print(list(G.neighbors(1))) # nodes that share an edge with edge 1
print(G.degree[1]) # num. nodes that share an edge with node 1
print(G.edges([2, 'm'])) # list of nodes that share edges with 2 and m
print(G.degree([2, 3])) # num of nodes that share edges with 2 and 3
```

```python
# removing nodes and edges
G.remove_node(2)
G.remove_nodes_from("spam")
G.remove_edge(1, 3)
```

```python
# accessing node/edge attributes
G.clear()
G = nx.Graph([(1, 2, {"color": "yellow"})])
print(G[1])
print(G[1][2])
print(G.edges[1, 2])
G.add_edge(1, 3)
G[1][3]['color'] = "blue"
G.edges[1, 2]['color'] = "red"
print(G.edges[1, 2])

# create a new graph with weighted edges and iterate through them
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in FG.adj.items():
   for nbr, eattr in nbrs.items():
       wt = eattr['weight']
       if wt < 0.5: print(f"({n}, {nbr}, {wt:.3})")
```
